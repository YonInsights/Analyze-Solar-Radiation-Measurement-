```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Load datasets
benin = pd.read_csv(r'D:\10 Academy_Kefya\Week 0\Analyze-Solar-Radiation-Measurement-\data\benin-malanville.csv')
sierra_leone = pd.read_csv(r'D:\10 Academy_Kefya\Week 0\Analyze-Solar-Radiation-Measurement-\data\sierraleone-bumbuna.csv')
togo = pd.read_csv(r'D:\10 Academy_Kefya\Week 0\Analyze-Solar-Radiation-Measurement-\data\togo-dapaong_qc.csv')


# Display the first 5 rows of each dataset
display(benin.head(5), sierra_leone.head(5), togo.head(5))
```

```python
# Function to remove 'Comments' column if it exists
def remove_comments_column(df):
    if 'Comments' in df.columns:
        df.pop('Comments')

# Apply the function to each DataFrame
remove_comments_column(benin)
remove_comments_column(sierra_leone)
remove_comments_column(togo)
```

```python
# Summary statistics for Benin
summary_statistics = benin.describe()
print(summary_statistics)

```

```python
# Summary statistics for sierraleone-bumbuna
summary_statistics = sierra_leone.describe()
print(summary_statistics)
```

```python
# Summary statistics for togo
summary_statistics = togo.describe()
print(summary_statistics)
```

```python
# Check for missing values
print("Missing values in Benin dataset:\n", benin.isnull().sum())
print("Missing values in Sierra Leone dataset:\n", sierra_leone.isnull().sum())
print("Missing values in Togo dataset:\n", togo.isnull().sum())

# Check for outliers using Z-score
from scipy.stats import zscore

benin_zscores = np.abs(zscore(benin.select_dtypes(include=[np.number])))
sierra_leone_zscores = np.abs(zscore(sierra_leone.select_dtypes(include=[np.number])))
togo_zscores = np.abs(zscore(togo.select_dtypes(include=[np.number])))

print("Outliers in Benin dataset:\n", (benin_zscores > 3).sum())
print("Outliers in Sierra Leone dataset:\n", (sierra_leone_zscores > 3).sum())
print("Outliers in Togo dataset:\n", (togo_zscores > 3).sum())
```

```python
#Convert Timestamp to Datetime
benin['Timestamp'] = pd.to_datetime(benin['Timestamp'])
sierra_leone['Timestamp'] = pd.to_datetime(sierra_leone['Timestamp'])
togo['Timestamp'] = pd.to_datetime(togo['Timestamp'])
```

```python
#Set Timestamp as Index
benin.set_index('Timestamp', inplace=True)
sierra_leone.set_index('Timestamp', inplace=True)
togo.set_index('Timestamp', inplace=True)
```

```python
#Plot Time Series Data
plt.figure(figsize=(14, 7))
plt.plot(benin['GHI'], label='GHI')
plt.title('GHI Over Time in Benin')
plt.xlabel('Time')
plt.ylabel('GHI')
plt.legend()
plt.show()
```

```python
#Plot Multiple Variables
plt.figure(figsize=(14, 7))
plt.plot(benin['GHI'], label='GHI')
plt.plot(benin['DNI'], label='DNI')
plt.plot(benin['DHI'], label='DHI')
plt.plot(benin['Tamb'], label='Tamb')
plt.title('Solar Irradiance and Temperature Over Time in Benin')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#Monthly Patterns benin
benin_monthly = benin.resample('M').mean()

plt.figure(figsize=(14, 7))
plt.plot(benin_monthly['GHI'], label='GHI')
plt.plot(benin_monthly['DNI'], label='DNI')
plt.plot(benin_monthly['DHI'], label='DHI')
plt.plot(benin_monthly['Tamb'], label='Tamb')
plt.title('Monthly Solar Irradiance and Temperature in Benin')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```
